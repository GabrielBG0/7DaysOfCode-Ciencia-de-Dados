```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('despesa_ceaps_2022.csv', sep=';', decimal=',', encoding='utf-8')
df
```

```python
df.describe()
```

```python
df_num = df[['ANO', 'MES', 'VALOR_REEMBOLSADO']]

for i in df_num.columns:
    plt.hist(df_num[i])
    plt.title(i)
    plt.show()
```

```python
df_cat = df[['SENADOR','TIPO_DESPESA','CNPJ_CPF','FORNECEDOR','DETALHAMENTO']]
for i in df_cat.columns:
    sns.barplot(df_cat[i].value_counts().index,
                df_cat[i].value_counts()).set_title(i)
    plt.show()
```

```python
df = df.drop(columns=['DETALHAMENTO'])
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoded = encoder.fit_transform(df['TIPO_DESPESA'].values)

df['TIPO_DESPESA'] = encoded
df
```

```python
sns.barplot(df['TIPO_DESPESA'].value_counts().index,
            df_cat['TIPO_DESPESA'].value_counts()).set_title('TIPO_DESPESA')
plt.show()
```

```python
for i in range(6):
    print('label ' + str(i) + ': ' + str(encoder.inverse_transform([i])))
```
