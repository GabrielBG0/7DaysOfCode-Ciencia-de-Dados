```python
import pandas as pd
df = pd.read_csv('despesa_ceaps_2019.csv', sep=';', decimal=',', encoding='utf-8')
df = pd.concat([df, pd.read_csv('despesa_ceaps_2020.csv', sep=';', decimal=',', encoding='utf-8')])
df = pd.concat([df, pd.read_csv('despesa_ceaps_2021.csv', sep=';', decimal=',', encoding='utf-8')])
df = pd.concat([df, pd.read_csv('despesa_ceaps_2022.csv', sep=';', decimal=',', encoding='utf-8')])
df
```

```python
df = df.drop(columns=['DETALHAMENTO'])
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoded = encoder.fit_transform(df['TIPO_DESPESA'].values)

df['TIPO_DESPESA'] = encoded
df
```

```python
for i in range(6):
    print('label ' + str(i) + ': ' + str(encoder.inverse_transform([i])))
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

df_valXtype = df[['VALOR_REEMBOLSADO', 'TIPO_DESPESA', 'MES']]
print(df_valXtype.corr())
sns.heatmap(df_valXtype.corr())
```

```python
print(df['SENADOR'].mode())
df_senador = df[df['SENADOR'] == 'CARLOS VIANA']
print(df_senador.describe())
df_senador
```

```python
print(df_senador['TIPO_DESPESA'].mode())
print(df_senador['FORNECEDOR'].mode())
df_senador['VALOR_REEMBOLSADO'].values.sum()
```

```python
sns.barplot(df_senador['TIPO_DESPESA'].value_counts().index,
            df_senador['TIPO_DESPESA'].value_counts()).set_title('TIPO_DESPESA')
plt.show()
```

```python
print(df_senador.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum())
sns.barplot(df_senador['TIPO_DESPESA'].value_counts().index.sort_values(),
            df_senador.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum()).set_title('VALOR_REEMBOLSADO')
plt.show()
```
